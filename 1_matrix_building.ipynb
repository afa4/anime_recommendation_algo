### Cleaning

```python
ratings = ratings[['username', 'anime_id', 'my_score']]
animes = animes[['anime_id', 'title', 'title_english', 'image_url', 'score', 'rank', 'genre']]
```

```python
ratings.head()
```

```python
animes.head()
```

### Anime x UserRating Matrix Building

```python
ratings_merged = animes.merge(ratings, on='anime_id')
ratings_merged.head()
```

```python
len(ratings_merged.index) # before 
```

```python
animes_rated_by_users = pd.pivot_table(ratings_merged, index='title', columns='username', values='my_score').fillna(0)
```

```python
animes_rated_by_users.head()
```

```python
# export dataframe animes_rated_by_users to csv
animes_rated_by_users.to_parquet('../data/animes_rated_by_users.parquet')
```
