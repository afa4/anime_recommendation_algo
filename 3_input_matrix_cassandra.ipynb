```python
from cassandra.cluster import Cluster
from ssl import SSLContext, PROTOCOL_TLSv1_2 , CERT_REQUIRED
from cassandra.auth import PlainTextAuthProvider
import os

ssl_context = SSLContext(PROTOCOL_TLSv1_2 )
ssl_context.load_verify_locations('./sf-class2-root.crt')
ssl_context.verify_mode = CERT_REQUIRED
credentials = session.get_credentials()
auth_provider = PlainTextAuthProvider(username=os.environ['CASSANDRA_USER'], password=os.environ['CASSANDRA_PASS'])
cluster = Cluster(['cassandra.us-east-1.amazonaws.com'], ssl_context=ssl_context, auth_provider=auth_provider, port=9142)
session = cluster.connect()
r = session.execute('select * from anime_keyspace.anime_similarity_cosine_matrix')
print(r.current_rows)
```
